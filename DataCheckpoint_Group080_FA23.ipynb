## Flight Prices

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("itineraries_random_2M-Copy1.csv")
df.head()
```

```python
# List of columns to keep
selected_columns = [
    'legId',
    'flightDate',
    'startingAirport',
    'destinationAirport',
    'isBasicEconomy',
    'isNonStop',
    'baseFare',
    'totalFare',
    'seatsRemaining',
    'segmentsAirlineName',
    'segmentsEquipmentDescription',
    'segmentsDurationInSeconds'
]
```

```python
# Creating a new DataFrame with only the selected columns
cleaned_column_df = df[selected_columns]
```

```python
# Dropping rows with missing values
cleaned_column_df = cleaned_column_df.dropna()
```

```python
# Renaming the columns
column_mapping = {
    'legId': 'Leg ID',
    'flightDate': 'Flight Date',
    'startingAirport': 'Starting Airport',
    'destinationAirport': 'Destination Airport',
    'isBasicEconomy': 'Basic Economy',
    'isNonStop': 'Non-Stop',
    'baseFare': 'Base Fare',
    'totalFare': 'Total Fare',
    'seatsRemaining': 'Seats Remaining',
    'segmentsAirlineName': 'Airline Name',
    'segmentsEquipmentDescription': 'Equipment Description',
    'segmentsDurationInSeconds': 'Duration in seconds'
}

cleaned_column_names_df = cleaned_column_df.rename(columns=column_mapping)
```

```python
# Modifying data stored in the "Airline Name" column
cleaned_column_names_df['Airline Name'] = cleaned_column_names_df['Airline Name'].apply(lambda s : s.split('||')[0])
```

```python
# Modifying data stored in the "Equipment Description" column
cleaned_column_names_df['Equipment Description'] = cleaned_column_names_df['Equipment Description'].apply(lambda s : str(s).split('||')[-1])
cleaned_column_names_df['Equipment Description'] = cleaned_column_names_df['Equipment Description'].str.lower().str.capitalize()
```

```python
# Modifying data stored in the "Duration in seconds" column
cleaned_column_names_df['Duration in seconds'] = cleaned_column_names_df['Duration in seconds'].apply(lambda s: sum(int(value) for value in str(s).split('||') if value.isdigit()))
```

```python
# Renaming the dataframe
df_prices = cleaned_column_names_df
```

```python
df_prices.describe()
```

```python
df_prices.dtypes
```

```python
df_prices.head()
```

```python
# Cheking for different aircrafts in the dataset
df_prices.groupby('Equipment Description').count()
```

```python
# Renaming redundant aircraft names and dropping "Boeing 787" due to unclear description
df_prices['Equipment Description'] = df_prices['Equipment Description'].replace({
    'Airbus industrie a320 sharklets': 'Airbus a320',
    'Airbus industrie a321 sharklets': 'Airbus a321',
    'Embraer 175 (enhanced winglets)': 'Embraer 175'
})
df_prices = df_prices[df_prices['Equipment Description'] != 'Boeing 787']
```

```python
# Dropping empty inputs/ NA
df_prices['Equipment Description'] = df_prices['Equipment Description'].replace('', pd.NA)
df_prices = df_prices.dropna(subset=['Equipment Description'])
df_prices.head()
```
